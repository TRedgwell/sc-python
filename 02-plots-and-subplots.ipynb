## Putting it all together 

```python
filenames = sorted(glob.glob('data/weather*.csv'))
filenames = filenames[0:3]

#If want to run on all then comment out the line above.

# overwrites filenames with the first 3 of the sorted weather dataset.
# f is a variable, so for every variable in filenames do this.
for f in filenames:
    print (f)
    
    data = numpy.loadtxt(fname=f, delimiter = ',')
    
    if numpy.max (data, axis=0)[0] ==0 and numpy.max (data, axis=0)[20] ==20:
        print ("suspicious looking maxima")
    elif numpy.sum(numpy.min(data, axis=0)) ==0:
        print ("minima add up to zero")
    else:
        print ("data looks ok")


    # Create a wide figure to hold the subplots.
    fig = matplotlib.pyplot.figure (figsize = (10.0, 3.0))

    # Create placeholders for plots
    subplot1 = fig.add_subplot (1,3,1)
    subplot2 = fig.add_subplot (1,3,2)
    subplot3 = fig.add_subplot (1,3,3)

    subplot1.set_ylabel('average')
    subplot1.plot(numpy.mean(data, axis=0))

    subplot2.set_ylabel('max')
    subplot2.plot(numpy.max(data, axis=0))

    subplot3.set_ylabel('min')
    subplot3.plot(numpy.min(data, axis=0))

    fig.tight_layout()
    matplotlib.pyplot.show()
```

## Functions

```python
def fahr_to_kelvin(temp):
    return ((temp - 32) * (5/9) + 273.15)
```

```python
print ('Freezing point of water: ', fahr_to_kelvin(32))
```

```python
print ('boiling point of water: ', fahr_to_kelvin(212))
```

```python
def analyse (filename):
    """To plot linegraphs of mean, min, and maximal data.
    
    This function reads a file (filename argument) and produces a plot containg 3 subplots, representing the average of 
    the data, the max, and the min.
    This function produces an image.
    """
    data = numpy.loadtxt(fname=filename, delimiter = ',')
    
    
    # Create a wide figure to hold the subplots.
    fig = matplotlib.pyplot.figure (figsize = (10.0, 3.0))

    # Create placeholders for plots
    subplot1 = fig.add_subplot (1,3,1)
    subplot2 = fig.add_subplot (1,3,2)
    subplot3 = fig.add_subplot (1,3,3)

    subplot1.set_ylabel('average')
    subplot1.plot(numpy.mean(data, axis=0))

    subplot2.set_ylabel('max')
    subplot2.plot(numpy.max(data, axis=0))

    subplot3.set_ylabel('min')
    subplot3.plot(numpy.min(data, axis=0))

    fig.tight_layout()
    matplotlib.pyplot.show()
```
